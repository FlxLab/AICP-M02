## 8. Model evaluation: Accuracy vs Overfitting Risk

Why: To identify the model that balances both high accuracy and low train/test loss difference, ensuring good performance and generalization.
Plot:  Scatter plot showing final accuracy vs. train/test loss difference for different models.

 Calculate Final Loss Differences

```python
# Calculate the final loss difference for each model
baseline_final_loss_diff = baseline_train_losses[-1] - baseline_test_losses[-1]
testB_final_loss_diff = testB_train_losses[-1] - testB_test_losses[-1]
testC_final_loss_diff = testC_train_losses[-1] - testC_test_losses[-1]
#testD_final_loss_diff = testD_train_losses[-1] - testD_test_losses[-1]
testE_final_loss_diff = testE_train_losses[-1] - testE_test_losses[-1]
testF_final_loss_diff = testF_train_losses[-1] - testF_test_losses[-1]
#testG_final_loss_diff = testG_train_losses[-1] - testG_test_losses[-1]
testH_final_loss_diff = testH_train_losses[-1] - testH_test_losses[-1]
testI_final_loss_diff = testI_train_losses[-1] - testI_test_losses[-1]  # Corrected

# List of final loss differences
final_loss_diffs = [baseline_final_loss_diff, testB_final_loss_diff, testC_final_loss_diff, 
                    testE_final_loss_diff, testF_final_loss_diff, 
                    testH_final_loss_diff, testI_final_loss_diff]
```

Create a Scatter Plot Comparing Final Accuracy and Loss Differences

```python
import matplotlib.pyplot as plt

# List of model names and their final accuracies
# model_names = ['Baseline', 'Test B', 'Test C', 'Test D', 'Test E', 'Test F', 'Test G', 'Test H','Test I']
# final_accuracies = [baseline_final_accuracy, testB_final_accuracy, testC_final_accuracy, testD_final_accuracy, 
#                     testE_final_accuracy, testF_final_accuracy, testG_final_accuracy, testH_final_accuracy,testI_final_accuracy]

model_names = ['Baseline', ' On the fly',' Shift', ' Regularization', ' Flip', ' 3 Lay',' 4 Lay']
final_accuracies = [baseline_final_accuracy, testB_final_accuracy, testC_final_accuracy, testE_final_accuracy, testF_final_accuracy, testH_final_accuracy, testI_final_accuracy]

# Scatter plot of final accuracy (x-axis) vs final loss difference (y-axis)
plt.figure(figsize=(8, 5))
plt.scatter(final_accuracies, final_loss_diffs)

# Annotate each point with the corresponding model name
for i, model_name in enumerate(model_names):
    plt.annotate(model_name, (final_accuracies[i], final_loss_diffs[i]))

# Add labels and title
plt.xlabel('Final Accuracy')
plt.ylabel('Final Train-Test Loss Difference')
plt.title('Comparison of Models: Accuracy vs Train-Test Loss Difference')
plt.grid(True)
plt.show()
```
